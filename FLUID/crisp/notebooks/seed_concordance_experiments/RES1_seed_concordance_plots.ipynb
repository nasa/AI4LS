```python
import os
import json
import pandas as pd
if os.getcwd() != '/home/jupyter/crisp':
    os.chdir("..")
%pwd
from nasa_mice.functions import *
from tqdm import tqdm

import matplotlib.pyplot as plt
import seaborn as sns
```

```python
metrics = ['top_10_overlap_percent','top_20_overlap_percent','top_50_overlap_percent','ranked_bias_overlap','cosine_similarity']
```

```python
fixed_data_model_seed_files = [f for f in os.listdir('results/fixed_data_seed') if '.' not in f]

fixed_data_model_seed_baseline = 'model_seed_0'
with open('results/fixed_data_seed/'+fixed_data_model_seed_baseline +'/results_for_bucket.json') as json_file:
    base = json.load(json_file)
    
fixed_data_model_seed_comparasons = {}
for c in tqdm(fixed_data_model_seed_files):
    with open('results/fixed_data_seed/'+c+'/results_for_bucket.json') as json_file:
        case = json.load(json_file)
        fixed_data_model_seed_comparasons[c] = data_comparisons(base, case, 0)
        
```

```python
fig, axs = plt.subplots(1, len(metrics), figsize=(20,5))
fig.suptitle('Linear IRM similarity for for varying model seeds')
idx=0
for metric in metrics:
    vals =  [fixed_data_model_seed_comparasons[c].metrics[metric] for c in fixed_data_model_seed_files]
    chart = sns.barplot(x=fixed_data_model_seed_files, y = vals, ax = axs[idx])
    chart.set_xticklabels(chart.get_xticklabels(), rotation=45, horizontalalignment='right')
    chart.set_title(metric)
    idx += 1
```

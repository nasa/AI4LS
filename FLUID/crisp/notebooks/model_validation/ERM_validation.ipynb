```python
import os
import numpy as np
os.chdir("../..")

from utils.get_synthetic_data_for_validation import get_datasets
from utils.vm_helpers import save_dict_to_json
from utils.gcp_helpers import save_json_to_bucket
from models.EmpericalRiskMinimization import EmpericalRiskMinimization

%pwd
```

```python
ERM_args = {
            "verbose": 0,
            "method": 'Linear',
            "cuda": False,
            "seed": 12,
            "epochs": 1,
            "hidden_dim": 256
        }
```

```python
examples = ["Example1", "Example2", "Example3", "Example4", "Example_Confounder", "Example_Nonlinear"]
n_seeds = 10
for example in examples:            
    results = {}
            
    for seed in range(n_seeds):
        print(example, ' seed ', int(seed))
        environment_datasets, val_dataset, test_dataset, config =  get_datasets(example, seed=seed)
        ERM_args["target"] = config["data_options"]["targets"]
        ERM_args["output_data_regime"] = config["data_options"]["output_data_regime"]
        ERM_args["columns"] = config["data_options"]["predictors"]

        erm = EmpericalRiskMinimization(environment_datasets, val_dataset, test_dataset, ERM_args)
        results[seed] = erm.results()
        results[seed]["seed"] = seed
            
    result_file = "results/validation/ERM_" + example + ".json"

    for seed in range(n_seeds):
#         results.get(seed).get("to_bucket")["test_acc"] = results.get(seed).get("to_bucket")["test_acc"].item()
#         results.get(seed)["test_acc"] = results.get(seed)["test_acc"].item()
        results.get(seed)["test_probs"] = list(np.float64(results.get(seed)["test_probs"]))
        results.get(seed)["test_labels"] = list(np.float64(results.get(seed)["test_labels"]))
        results.get(seed)["feature_coeffients"] = list(np.float64(results.get(seed)["feature_coeffients"]))

    save_dict_to_json(results, result_file)
```
